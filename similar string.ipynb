## Similar Substrings

### Problem Description

You are given two strings, A and B. Two strings, X and Y, are said to be similar if :

Length(X) = Length(Y).

The frequency of each character in string X should be equal to that of each character in string Y.

String B consists of only lowercase Latin characters, and String A consists of lowercase Latin letters and characters “?”. You can replace the “?” character with any lowercase Latin letter.

Find the substrings of string A similar to string B.

For ex: “a?b?c” is similar to “aabcc”, “bcabb”, etc.



### Problem Constraints

1 ≤ |A| ≤ 105

1 ≤ |B| ≤ 105



### Input Format

The first argument given is a string, A.

The second argument given is a string, B.



### Output Format

Return  substrings of string A that are similar to string B.



#### Example Input

Input 1:

 A = "bb??x???"      B = "aab"

Input 2:

 A = "ab?c"          B = "acb"



Example Output

Output 1:

 ['baa','aab']

Output 2:

 ['bac','acb']



Example Explanation

Explanation 1:

We can find two similar substrings in A:

1) "b??": we can replace both the "?" with "a", to form "baa" which is similar to "aab".

2) "???": we can replace it with "aab".

Explanation 2:

We can find two similar substrings in A:

1) "ab?": we can replace the "?" with "c", to form "abc" which is similar to "acb".

2) "b?c": we can replace the "?" with "a", to form "bac" which is similar to "acb".

```python
def similar_substrings(A, B):
    n, m = len(A), len(B)
    if m > n:
        return []  # No possible substrings if B is longer than A

    # Function to build frequency count of a string
    def build_freq(s):
        freq = {}
        for char in s:
            if char in freq:
                freq[char] += 1
            else:
                freq[char] = 1
        return freq

    # Frequency counter for B
    freq_b = build_freq(B)

    # Sliding window counters for A
    current_freq = {}
    question_marks = 0
    result = []

    # Initialize the frequency count for the first window
    for i in range(m):
        if A[i] == '?':
            question_marks += 1
        else:
            if A[i] in current_freq:
                current_freq[A[i]] += 1
            else:
                current_freq[A[i]] = 1

    # Function to check if the current window can be transformed to match B
    def is_match():
        for char in freq_b:
            needed = freq_b[char]
            has = current_freq.get(char, 0)
            if has + question_marks < needed:
                return False
        total_needed = sum(freq_b.values())
        total_has = sum(current_freq.values()) + question_marks
        return total_has >= total_needed

    # Check the first window explicitly
    if is_match():
        result.append(A[:m].replace('?', 'a'))  # Example replacement (illustrative)

    # Sliding over the rest of A
    for i in range(m, n):
        start_char = A[i - m]
        end_char = A[i]

        # Remove the start character from the window
        if start_char == '?':
            question_marks -= 1
        else:
            if current_freq[start_char] == 1:
                del current_freq[start_char]
            else:
                current_freq[start_char] -= 1

        # Add the new character to the window
        if end_char == '?':
            question_marks += 1
        else:
            if end_char in current_freq:
                current_freq[end_char] += 1
            else:
                current_freq[end_char] = 1

        # Check the current window
        if is_match():
            result.append(A[i - m + 1:i + 1].replace('?', 'a'))  # Example replacement (illustrative)

    return result

# Test cases
print(similar_substrings("bb??x???", "aab"))
print(similar_substrings("ab?c", "acb"))
```

```python
class Solution:
    def solve(self, A, B):
        def count_freq(s):
            freq = [0] * 26
            wildcards = 0
            for char in s:
                if char == '?':
                    wildcards += 1
                else:
                    freq[ord(char) - ord('a')] += 1
            return freq, wildcards

        # Initialize variables
        n = len(A)
        m = len(B)
        if m > n:
            return 0
        
        target_freq, _ = count_freq(B)
        current_freq, wildcards = count_freq(A[:m])
        count = 0

        # Function to compare frequencies considering wildcards
        def is_matching(target_freq, current_freq, wildcards):
            required_wildcards = 0
            for i in range(26):
                if target_freq[i] > current_freq[i]:
                    required_wildcards += (target_freq[i] - current_freq[i])
            return required_wildcards <= wildcards

        # Check if the substring starting from index 0 is similar to B
        if is_matching(target_freq, current_freq, wildcards):
            count += 1

        # Slide the window and check each substring
        for i in range(1, n - m + 1):
            left_char = A[i - 1]
            right_char = A[i + m - 1]

            # Update wildcard count and frequencies
            if left_char == '?':
                wildcards -= 1
            else:
                current_freq[ord(left_char) - ord('a')] -= 1
            
            if right_char == '?':
                wildcards += 1
            else:
                current_freq[ord(right_char) - ord('a')] += 1

            # Check if the substring is similar to B
            if is_matching(target_freq, current_freq, wildcards):
                count += 1

        return count
```

```python
def count_chars(s):
    count_dict = {}
    for ch in s:
        if ch != '?':
            count_dict[ch] = count_dict.get(ch, 0) + 1
    return count_dict

def similar_substrings(x, y):
    len_x, len_y = len(x), len(y)
    if len_x < len_y:
        return []  # No substrings of sufficient length

    y_counts = count_chars(y)
    result_count = 0
    results_arr = []
    
    # Prepare for the first window
    current_window_count = count_chars(x[:len_y])
    question_marks = x[:len_y].count('?')
    
    def can_form_y():
        if len_x < len_y:
            return False
        needed_changes = 0
        for char in y_counts:
            if y_counts[char] > current_window_count.get(char, 0):
                needed_changes += y_counts[char] - current_window_count.get(char, 0)
        return needed_changes <= question_marks
    
    # Check the first window
    if can_form_y():
        result_count += 1
        results_arr.append(x[:len_y])

    # Move the window across x
    for i in range(1, len_x - len_y + 1):
        old_char = x[i - 1]
        new_char = x[i + len_y - 1]
        
        if old_char == '?':
            question_marks -= 1
        else:
            if current_window_count[old_char] == 1:
                del current_window_count[old_char]
            else:
                current_window_count[old_char] -= 1
        
        if new_char == '?':
            question_marks += 1
        else:
            current_window_count[new_char] = current_window_count.get(new_char, 0) + 1

        # Check this new window
        if can_form_y():
            result_count += 1
            results_arr.append(x[i:i+len_y])

    return results_arr

# Example usage:
x = 'bb??x???'
y = 'aab'
print(similar_substrings(x, y))
```

```python
def similar_substrings(a, b):
    lenA, lenB = len(a), len(b)
    if lenA < lenB:
        return 0  # Impossible to have a valid window if a is shorter than b

    # Frequency dictionary for b
    freqB = {}
    for ch in b:
        freqB[ch] = freqB.get(ch, 0) + 1

    # Current frequency window in a
    freqA = {}
    question_marks = 0  # To count how many '?' are there in the current window
    result_count = 0
    
    # Initialize the first window
    for i in range(lenB):
        ch = a[i]
        if ch == '?':
            question_marks += 1
        else:
            freqA[ch] = freqA.get(ch, 0) + 1

    # Function to check if current window can be transformed to b
    def is_valid():
        required_changes = 0
        for ch in freqB:
            count_in_B = freqB[ch]
            count_in_A = freqA.get(ch, 0)
            if count_in_A < count_in_B:
                required_changes += count_in_B - count_in_A
        return required_changes <= question_marks

    # Check the first window
    if is_valid():
        result_count += 1

    # Slide the window across the rest of a
    for i in range(lenB, lenA):
        new_char = a[i]
        old_char = a[i - lenB]

        # Remove the influence of the old_char from the window
        if old_char == '?':
            question_marks -= 1
        else:
            freqA[old_char] -= 1
            if freqA[old_char] == 0:
                del freqA[old_char]

        # Add the influence of the new_char to the window
        if new_char == '?':
            question_marks += 1
        else:
            freqA[new_char] = freqA.get(new_char, 0) + 1

        # Check this new window
        if is_valid():
            result_count += 1

    return result_count

# Example usage:
print(similar_substrings("bb??x???", "aab"))  # Output: 2
print(similar_substrings("ab?c", "acb"))      # Output: 2
```
